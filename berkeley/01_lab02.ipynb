```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

temps = np.array([26.6 , 38.5 , 34.4 , 34 , 31 , 23.6])

# plot histogram
plt.hist(temps, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Histogram of Temperature')
plt.show()
```

- Prior: Probability distribution that represents the initial beliefs
- The true average temperature is a parameter Œ∏.
- The uncertainty or spread in temperature is captured by standard deviation œÉ.

We want to estimate Œ∏ (mean) and œÉ (standard deviation) from this data using Bayesian inference.

## Defining Priors

We have no idea what Œ∏ or œÉ could be - they could be almost anything within reason.

So we choose **uninformative priors**:

$$
\theta \sim \text{Unif}(-C, C) \quad \text{and} \quad \log \sigma \sim \text{Unif}(-C, C)
$$

This means:

* Œ∏ is equally likely to be any value in $[-C, C]$
* log œÉ is equally likely to be any value in $[-C, C]$

This is a **flat prior** over a wide plausible range.

---

## üìê Mathematical Form

Prior density function:

$$
f_{\theta, \sigma}(\theta, \sigma) = \frac{I\{-C < \theta < C,\; -C < \log \sigma < C\}}{4\sigma C^2}
$$

* $I\{\cdot\}$ is an indicator function: 1 if the condition is true, 0 otherwise.
* The $\frac{1}{\sigma}$ term comes from the transformation of log œÉ ‚Üí œÉ.
* This ensures the prior integrates to 1 (proper density).




```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Priors
C = 100  # Large constant for "uninformative" prior

n_samples = 10000

# Œ∏ ~ Uniform(-C, C)
theta_prior = np.random.uniform(-C, C, n_samples)

# log(œÉ) ~ Uniform(-C, C) => œÉ = exp(logœÉ)
log_sigma_prior = np.random.uniform(-C, C, n_samples)
sigma_prior = np.exp(log_sigma_prior)

# Plot priors
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(theta_prior, bins=50, color='lightcoral', edgecolor='black')
plt.title('Prior Distribution of Œ∏ (Mean)')
plt.xlabel('Œ∏')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(sigma_prior, bins=50, color='lightgreen', edgecolor='black')
plt.title('Prior Distribution of œÉ (Std Dev)')
plt.xlabel('œÉ')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
