Linear Interpolation

$f(x) = f(x_0) + \frac{f(x_1) - f(x_0)}{x_1 - x_0}(x - x_0)$


Nearest Interpolation

$f(x) = f(x_i) \quad \text{where} \quad x_i = \operatorname{argmin}_{x_k \in \text{known}} |x_k - x|$


Spline Interpolation

$f_i(x) = a_i + b_i(x - x_i) + c_i(x - x_i)^2$ 

Polynomial Interpolation

$f(x) = a_0 + a_1x + a_2x^2 + a_3x^3 + a_4x^4 + a_5x^5$

## 📐 1. **Wide Format - Not recommended**

### 💡 Definition:

* Each **row = one time point**
* Each **column = one time series**
* **Date** is in the index or a separate column
* Most common in spreadsheets, classical time series tools (like Excel or statsmodels)

### 🧾 Example:

| datetime   | sensor\_A | sensor\_B | sensor\_C |
| ---------- | --------- | --------- | --------- |
| 2025-06-21 | 3.1       | 5.2       | 8.3       |
| 2025-06-22 | 3.3       | 5.1       | 8.5       |
| 2025-06-23 | 3.2       | 5.3       | 8.4       |



## 🔁 2. **Expanded Format (aka Long Format)**

### 💡 Definition:

* Each **row = one observation**
* All time series go into **a single column**, with a separate identifier for which series it is
* More aligned with **relational databases** and tidy data principles

### 🧾 Example:

| datetime   | sensor\_id | value |
| ---------- | ---------- | ----- |
| 2025-06-21 | A          | 3.1   |
| 2025-06-21 | B          | 5.2   |
| 2025-06-21 | C          | 8.3   |
| 2025-06-22 | A          | 3.3   |
| 2025-06-22 | B          | 5.1   |
| 2025-06-22 | C          | 8.5   |


## 🧳 3. **Compact Format**

### 💡 Definition:

* Each **row = one full time series**
* The **time dimension is stored in a list/array**
* Needs two extra columns: `start_datetime` and `frequency`

### 🧾 Example:

| sensor\_id | start\_datetime     | freq | values           |
| ---------- | ------------------- | ---- | ---------------- |
| A          | 2025-06-21 00:00:00 | 1D   | \[3.1, 3.3, 3.2] |
| B          | 2025-06-21 00:00:00 | 1D   | \[5.2, 5.1, 5.3] |
| C          | 2025-06-21 00:00:00 | 1D   | \[8.3, 8.5, 8.4] |



## 🧭 Summary Table

| Feature        | Wide Format   | Expanded Format           | Compact Format                      |
| -------------- | ------------- | ------------------------- | ----------------------------------- |
| Rows           | = time points | = observations            | = time series                       |
| Columns        | each series   | `id`, `datetime`, `value` | `id`, `values`, `start`, `freq`     |
| Metadata       | Not supported | Easy to include           | Easy to include                     |
| Time as        | Index/column  | Column                    | Implicit (via array + start + freq) |
| Scaling        | Poor          | Excellent                 | Excellent                           |
| Works with ML  | No            | Yes                       | Yes                                 |
| Works with SQL | No            | Yes                       | Some effort needed                  |



### 🔮 Best Format for Time Series Forecasting

The **best format** depends on your model and forecasting approach — but generally:


## ✅ **Best Overall: Expanded Format (a.k.a. Long Format)**

### Why?

* **Most ML/DL libraries** (like Prophet, XGBoost, PyTorch Forecasting, GluonTS, etc.) expect or prefer this format.
* **Scales easily** to many time series (multivariate or panel forecasting).
* **Metadata-friendly**: allows adding features like region, category, etc.
* **Compatible with time-varying features** (holidays, weather, lags, etc.).
* **Easy to batch and feed into models.**



### 🔁 Example: Expanded Format

| datetime   | series\_id | value | region | pricing\_type |
| ---------- | ---------- | ----- | ------ | ------------- |
| 2023-01-01 | A          | 123   | East   | Dynamic       |
| 2023-01-01 | B          | 55    | West   | Standard      |
| 2023-01-02 | A          | 140   | East   | Dynamic       |

> This can be directly used in:
>
> * Facebook Prophet (with grouping logic)
> * PyTorch Forecasting (`TimeSeriesDataSet`)
> * XGBoost/LightGBM (after feature engineering)


## ✅ **Compact Format** — for deep learning models with regular time intervals

### Use if:

* Your time series are **regularly sampled** (e.g., hourly/daily)
* You're training **RNNs, Transformers, or Temporal CNNs**
* You're using libraries like **tsai**, **Darts**, or **GluonTS**

It is **memory-efficient** and well-suited for:

* Storing long time series
* Quick batch training
* Reducing repeated timestamps

But not ideal for:

* Irregular sampling
* Missing timestamps
* External time-varying features


## ❌ **Wide Format** — not recommended

* Hard to scale with many series
* Not compatible with metadata or features
* Difficult to use with most ML/DL tools

Use it **only for EDA or classical univariate forecasting** (like ARIMA on one column).


## ✅ Conclusion

| Goal                      | Best Format         |
| ------------------------- | ------------------- |
| ML / Deep Learning Models | **Expanded (long)** |
| Memory-efficient storage  | **Compact**         |
| Simple ARIMA on 1 series  | **Wide**            |
